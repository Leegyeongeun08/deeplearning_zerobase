```python
lst_df=lst[['법령일련번호','법령명한글']]
```

```python
merged_df = pd.merge(main, lst, on='법령일련번호', how='left')
merged_df.head()
```

```python
merged_df.info()
```

```python
delete_keywords = ['평창','세월호','새만금', '전쟁','설립ㆍ운영','지정 및 운영','직무집행', '계엄','혁신도시',
                   '공무원','공직자','국가유공자','국영일','설치법','설립 및 운영','국방','군무원','공항','임용','군납',
                   '군용','군인','상호방위조약','독도','문화','무형','물가안정','민방위','병역','육성','설치','조성',
                   '운영','정부산하기관','텔레비전','지속가능','지역','천문','항공','항로','항만','해군','대회']
delpattern = '|'.join(delete_keywords)

# keyword Content
merged_df['keyword Contents'] = merged_df['Contents']

# 특정 키워드 삭제
filtered_df = merged_df[~merged_df['법령명한글'].str.contains(delpattern)]

# 1조, 2조 삭제
filtered_df = filtered_df[~filtered_df['keyword Contents'].str.contains('\(Purpose\)|\(Definitions\)')]

pd.set_option('display.max_colwidth', None)
filtered_df.head(2)
```

```python
# 패턴들을 공백으로 대체
pattern = (
    r'<(Amended on|Newly Inserted on) [A-Za-z]{3}\. \d{1,2}, \d{4}>'  # <Amended on Sep. 11, 2020> 형식 제거
    r'|<Amended on [^>]+>'
    r'|\[(This Article Newly Inserted on|Title Amended on) [A-Za-z]{3}\. \d{1,2}, \d{4}\]'  # [This Article Newly Inserted on Mar. 7, 2023] 형식 제거
    r'|\bArticle \d+'  # Article 28
    r'|Deleted [A-Za-z]{3} \d{4}'  # Deleted Aug 2007
    r'|\d{4} [A-Za-z]{3} \d{1,2}'
    r'|This Article Wholly Amended.*'  # This Article Wholly Amended
    r'|[A-Za-z]{3}\. \d{1,2}, \d{4}'  # Mar. 21, 2012 형식의 날짜 제거
    r'|[A-Za-z]{3} \d{1,2}, \d{4}'  # Mar 21, 2012 형식의 날짜 제거
    r'|\d{4} [A-Za-z]{3} \d{1,2}'  # 2012 Nov 19 형식의 날짜 제거
    r'|\d{1,2} [A-Za-z]{3} \d{4}'  # 21 Mar 2012 형식의 날짜 제거
)

filtered_df['keyword Contents'] = filtered_df['keyword Contents'].str.replace(pattern, '', regex=True)
filtered_df['keyword Contents'] = filtered_df['keyword Contents'].str.replace(r'\n', ' ', regex=True) # \n
# 확인
filtered_df.head(6)
```

```python
filtered_df.info()
```

```python
kw_model = KeyBERT('distilbert-base-nli-mean-tokens')
tqdm.pandas()

#  키워드 추출
def extract_top_keywords(doc):
    keywords = kw_model.extract_keywords(
        doc,
        keyphrase_ngram_range=(1, 3),
        stop_words='english',
        top_n=3,
        highlight=True,
        use_maxsum=True,
        use_mmr=True
    )
    return ', '.join([f"{item[0]} ({item[1]:.4f})" for item in keywords])

filtered_df['Top_3_Keywords'] = filtered_df['keyword Contents'][:6].progress_apply(extract_top_keywords)
```

```python
filtered_df[:6]
```

```python
kw_model = KeyBERT(model='distilbert-base-nli-mean-tokens')
# 적절한 n_gram 파악하기
test_df = filtered_df.sample(n=5000, random_state=1)

# n-gram
def test_ngrams(doc, ngram_range):
    keywords = kw_model.extract_keywords(
        doc,
        keyphrase_ngram_range=ngram_range,
        stop_words='english',
        top_n=3
    )
    return keywords

# n-gram 범위
ngram_ranges = [(1, 1), (1, 2), (1, 3)]

# 결과
results = {}


for ngram_range in ngram_ranges:
    test_df[f'Keywords_{ngram_range}'] = test_df['keyword Contents'].progress_apply(lambda x: test_ngrams(x, ngram_range))
    results[ngram_range] = test_df[f'Keywords_{ngram_range}']

# 출력
for ngram_range, keywords in results.items():
    print(f"n-gram range: {ngram_range}")
    print(keywords.head(20))
```

```python
kw_model = KeyBERT('distilbert-base-nli-mean-tokens')
tqdm.pandas()

#  키워드 추출
def extract_top_keywords(doc):
    keywords = kw_model.extract_keywords(
        doc,
        keyphrase_ngram_range=(2, 3),
        stop_words='english', # 불용어 제거
        top_n=3,
        diversity=0.5, # 다양성 비율
        use_maxsum=True, #최대합 기법 사용 여부(유사성 최소화하면서 중요한 키워드 선택하는 방법)
        use_mmr=True # 최대 한계적 비율(유사하지 않은 키워드 선택하는 방법)
    )
    return ', '.join([f"{item[0]} ({item[1]:.4f})" for item in keywords])

filtered_df['Top_3_Keywords'] = filtered_df['keyword Contents'].progress_apply(extract_top_keywords)
```
