## 경은님 코드에서 제외리스트(한글)을 GPT로 번역
- delete_keywords_english

```python
# KeyBERT 모델 로드
kw_model = KeyBERT('multi-qa-MiniLM-L6-cos-v1')
```

```python
# 제외 단어 리스트
delete_keywords_english = ['(Purpose)', '(Definitions)', 'PyeongChang Olympic', 'PyeongChang','Sewol','Saemangeum', 'war','establishment and operation','designation and operation','performance of duties', 'martial law','innovation city',
                   'public official','public servant','national merit','Kookyoungil','establishment law','establishment and operation','national defense','military official','airport','appointment','military supply',
                   'military use','soldier','mutual defense treaty','Dokdo','culture','intangible','price stabilization','civil defense','military service','fostering','installation','creation',
                   'operation','government affiliated organizations','television','sustainability','region','astronomy','aviation','airway','port','navy','competition']
exclude_chars = set(string.punctuation)
```

```python
# 텍스트에서 키워드 추출 함수 정의 (MMR 사용, n-gram 범위 설정)
# 단일 단어 키워드가 랜덤으로 추출되는 것이 아니라 텍스트의 문맥을 기반으로 가장 관련성 높은 단어가 선택
def extract_keywords(text, top_n=5): # 키워드 추출 5개로 변경함 / keyphrase_ngram_range 옵션을 (1, 1)으로 설정해 단일 단어 키워드를 추출
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 3), use_mmr=True, diversity=0.5, stop_words='english', top_n=top_n)
    filtered_keywords = [(kw, score) for kw, score in keywords if all(word not in delete_keywords_english for word in kw.split())]

    return filtered_keywords[:top_n]
```

```python
import logging
from tqdm import tqdm

# 로깅 설정
logging.basicConfig(level=logging.INFO, filename='keyword_extraction.log', filemode='w',
                    format='%(asctime)s - %(levelname)s - %(message)s')

# 진행 상황 바 설정
tqdm.pandas()

# 'Keywords' 컬럼 추가
df_main['Keywords'] = None

# 키워드 추출 및 데이터프레임 업데이트
for idx, row in tqdm(df_main.iterrows(), total=df_main.shape[0]):
    content = row['Contents']
    if pd.notnull(content):
        # 각 아티클의 내용에서 키워드 추출 (MMR 사용)
        keywords = extract_keywords(content, top_n=5)
        # 데이터프레임에 키워드 할당 (형식: 키워드 (점수), 키워드 (점수), ...)
        formatted_keywords = ', '.join([f"{kw} ({score:.4f})" for kw, score in keywords])
        df_main.at[idx, 'Keywords'] = formatted_keywords
        # 추출된 키워드의 개수와 키워드를 로깅
        logging.info(f"추출된 키워드 5개: {formatted_keywords}")

# 데이터프레임 이름 변경함
df_main_Keywords = df_main

# 저장
df_main_Keywords.to_csv('Keywords_multi_light.csv', encoding='euc-kr')

# 로깅 파일 생성 완료 메시지
logging.info('Keyword extraction and CSV file saving completed.')
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModel
from torch.utils.data import DataLoader, Dataset
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 데이터 불러오기
file_path = '/content/keywords_with_cluster.csv'
data = pd.read_csv(file_path)

# 토크나이저 및 모델 로드
tokenizer = AutoTokenizer.from_pretrained("law-ai/InLegalBERT")
model = AutoModel.from_pretrained("law-ai/InLegalBERT")
model.to('cuda' if torch.cuda.is_available() else 'cpu')

# 문장 목록 준비
sentence_list_A = data['extract_keywords'].apply(lambda x: x.replace(", ", " ")).tolist()
sentence_list_B = data['keyword Contents'].tolist()
```

```python
class SentenceDataset(Dataset):
    def __init__(self, sentences, tokenizer, max_length=128):
        self.sentences = sentences
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.sentences)

    def __getitem__(self, idx):
        sentence = self.sentences[idx]
        encoding = self.tokenizer(sentence, padding='max_length', truncation=True,
                                  max_length=self.max_length, return_tensors='pt')
        encoding = {key: val.squeeze(0) for key, val in encoding.items()}
        return encoding

```

```python
def embed_sentences_in_batches(sentences, model, tokenizer, batch_size=16):
    dataset = SentenceDataset(sentences, tokenizer)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    all_embeddings = []

    with torch.no_grad():
        for batch in dataloader:
            inputs = {key: val.to(model.device) for key, val in batch.items()}
            outputs = model(**inputs)
            embeddings = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
            all_embeddings.extend(embeddings)

    return np.array(all_embeddings)
```

```python
# 배치로 코사인 유사도 계산 및 매핑
def calculate_cosine_similarity_in_batches(embeddings_A, embeddings_B, batch_size=16):
    num_batches_A = (len(embeddings_A) + batch_size - 1) // batch_size
    num_batches_B = (len(embeddings_B) + batch_size - 1) // batch_size

    best_matches = []
    best_matches_scores = []

    for i in range(num_batches_A):
        start_A = i * batch_size
        end_A = min((i + 1) * batch_size, len(embeddings_A))

        batch_A = embeddings_A[start_A:end_A]

        batch_best_matches = []
        batch_best_matches_scores = []

        for j in range(num_batches_B):
            start_B = j * batch_size
            end_B = min((j + 1) * batch_size, len(embeddings_B))

            batch_B = embeddings_B[start_B:end_B]

            cosine_similarities = cosine_similarity(batch_A, batch_B)

            for k in range(len(batch_A)):
                best_match_idx = np.argmax(cosine_similarities[k])
                best_match_score = np.max(cosine_similarities[k])

                batch_best_matches.append(start_B + best_match_idx)
                batch_best_matches_scores.append(best_match_score)

        best_matches.extend(batch_best_matches)
        best_matches_scores.extend(batch_best_matches_scores)

    return best_matches, best_matches_scores
```

```python
# 배치로 코사인 유사도 계산 및 매핑 수행
batch_size = 64
best_matches, best_matches_scores = calculate_cosine_similarity_in_batches(embeddings_A, embeddings_B, batch_size)
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 토크나이저 및 모델 로드
tokenizer = AutoTokenizer.from_pretrained("law-ai/InLegalBERT")
model = AutoModelForSequenceClassification.from_pretrained("./fine-tuned-model")
model.to('cuda' if torch.cuda.is_available() else 'cpu')
model.eval()
```

```python
class SentenceDataset(Dataset):
    def __init__(self, sentences, tokenizer, max_length=128):
        self.sentences = sentences
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.sentences)

    def __getitem__(self, idx):
        sentence = self.sentences[idx]
        encoding = self.tokenizer(sentence, padding='max_length', truncation=True,
                                  max_length=self.max_length, return_tensors='pt')
        encoding = {key: val.squeeze(0) for key, val in encoding.items()}
        return encoding

def embed_sentences_in_batches(sentences, model, tokenizer, batch_size=16):
    dataset = SentenceDataset(sentences, tokenizer)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    all_embeddings = []

    with torch.no_grad():
        for batch in dataloader:
            inputs = {key: val.to(model.device) for key, val in batch.items()}
            outputs = model(**inputs)
            # 임베딩 벡터는 logits로 사용합니다.
            embeddings = outputs.logits.cpu().numpy()
            all_embeddings.extend(embeddings)

    return np.array(all_embeddings)
```

```python
# 데이터 불러오기
file_path = '/content/statute_main.csv'
data = pd.read_csv(file_path)
```

```python
articles = data['Contents'].tolist()
batch_size = 64
article_embeddings = embed_sentences_in_batches(articles, model, tokenizer, batch_size)
```

```python
def recommend_articles(user_input, articles, article_embeddings, model, tokenizer, top_k=5):
    user_embedding = embed_sentences_in_batches([user_input], model, tokenizer, batch_size=1)
    cosine_similarities = cosine_similarity(user_embedding, article_embeddings)
    best_matches = np.argsort(cosine_similarities[0])[::-1][:top_k]
    recommendations = [{'article': articles[idx], 'score': cosine_similarities[0][idx]} for idx in best_matches]
    return recommendations
```
